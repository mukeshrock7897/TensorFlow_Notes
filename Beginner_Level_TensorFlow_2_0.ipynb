##Basic Concepts

### **Tensors**


*  Tensor is a multi-dimensional array used to represent data. Tensors are the primary data structures that TensorFlow operates on


*  Tensor are identified by three parameter

  1.   Rank: Identify the number of dimensions of a tensor Ex- Scaler , Vector , Matrix
  2.   Shape: Number of rows and columns of tensor
  3.   Type: data types assigned to tensor's element Ex- integer , float , string , boolean


### Types of Tensors

* Scalar (0-D Tensor): A single number, e.g., 3.0.
* Vector (1-D Tensor): An array of numbers, e.g., [1.0, 2.0, 3.0].
* Matrix (2-D Tensor): A 2D array of numbers, e.g., [[1.0, 2.0], [3.0, 4.0]].
* Higher-Dimensional Tensors: Arrays with more than two dimensions.

### **Variable**

* variable is a special kind of tensor that represents a modifiable tensor.
Variables are primarily used to hold and update parameters in machine learning models during training. Unlike constants, the values of variables can be changed, which is essential for learning and optimization processes.

* To create a variable in TensorFlow, you use the tf.Variable class. Variables can be initialized with any tensor, and their values can be updated using various TensorFlow operations.

* In variable type and shape you can't change , only and only you change its elements(data)

## **Example of Creating and Using Variables**

```python
import tensorflow as tf

# Create a variable initialized to a scalar value
scalar_var = tf.Variable(3.0, dtype=tf.float32)
print("Initial Scalar Variable:", scalar_var.numpy())

# Create a variable initialized to a vector (1-D tensor)
vector_var = tf.Variable([1.0, 2.0, 3.0], dtype=tf.float32)
print("Initial Vector Variable:", vector_var.numpy())

# Create a variable initialized to a matrix (2-D tensor)
matrix_var = tf.Variable([[1.0, 2.0], [3.0, 4.0]], dtype=tf.float32)
print("Initial Matrix Variable:", matrix_var.numpy())

print("\n\n\n")
# Update the value of the scalar variable
scalar_var.assign(5.0)
print("Updated Scalar Variable:", scalar_var.numpy())

# Update the value of the vector variable
vector_var.assign([4.0, 5.0, 6.0])
print("Updated Vector Variable:", vector_var.numpy())

# Update a specific element of the matrix variable
matrix_var[0, 0].assign(7.0)
print("Updated Matrix Variable:", matrix_var.numpy())
```

### **Placeholder**

* Placeholder in TensorFlow is a variable that we will assign data to later
* It allows us to create our operations and build our computation graph, without needing the data.
* In TensorFlow terminology, we then feed data into the graph through these placeholders
* Placeholders are replaced by function arguments in tf.function-decorated functions.

## **Example**

```python
import tensorflow as tf

# Define a function that acts similarly to using placeholders
@tf.function
def compute(scalar, matrix):
    return scalar * matrix

# Define the input data
scalar_value = 2.0
matrix_value1 = tf.constant([[1.0, 2.0], [3.0, 4.0]], dtype=tf.float32)
matrix_value2 = tf.constant([[10.0, 20.0], [30.0, 40.0]], dtype=tf.float32)

# Call the function with the actual values
result = compute(scalar_value, matrix_value1)
print("Result:", result)

print("\n\n")

result = compute(scalar_value, matrix_value2)
print("Result:", result)
```

### **Sparse Tensor**

* sparse tensor in TensorFlow is a tensor that is primarily composed of zeros, with only a few non-zero elements
* Sparse tensors are efficient for representing data that contains many zero values, as they save memory and computational resources compared to dense tensors.
* There are two ways to create a sparse tensor.
*  **1- By Directly Specifying Indices and Values**
*  **2- Creating from Dense Tensor**



**1- By Directly Specifying Indices and Values**



**When you are creating a sparse tensor, you need to specify the following three components:**

* Values: These are the non-zero values, represented in 1D tensor.
* Indices: These are the indices of the non-zero values in the tensor, represented in 2D tensor.
* Dense Shape: It specifies the overall shape of the tensor in 1D tensor


```python
import tensorflow as tf

# Define non-zero values and their indices
values = tf.constant([41, 30], dtype=tf.int32)
indices = tf.constant([[1, 2], [3, 3]], dtype=tf.int64)
dense_shape = tf.constant([4, 8], dtype=tf.int64)

# Create a sparse tensor
sparse_tensor = tf.sparse.SparseTensor(indices=indices, values=values, dense_shape=dense_shape)

# Print the sparse tensor
print(sparse_tensor)
```

### **2- Creating from Dense Tensor**

* In case you are working on a large dataset, basically comprising all tensors. However, you notice that most of the values in the tensor are zero. Then, you can use tf.sparse module to work with sparse tensors. Here, you will have to choose tf.sparse.from_dense method.

```python
import tensorflow as tf

# Create a dense tensor
dense_tensor = tf.constant([[1, 0, 0], [0, 0, 2], [0, 3, 0]], dtype=tf.float32)

# Convert the dense tensor to a sparse tensor
sparse_tensor = tf.sparse.from_dense(dense_tensor)

# Print the dense and sparse tensors
print("Dense Tensor:")
print(dense_tensor)

print("\nSparse Tensor:")
print(sparse_tensor)
```

### Graph and Session
* graph and session were used in tensorflow 1.0
* since in tensorflow 2.0 these moved into one decorator @tf.function

**Graphs:**
* In TensorFlow 1.x, graphs were the core concept for defining computations. They served as a static representation of the computational flow, consisting of nodes (operations) and edges (tensors flowing between them). You'd manually create and manage these graphs.

**Sessions:**
* Sessions acted as execution environments for the graphs. You'd create a session, specify the graph to execute, feed data to placeholders (if any), and call sess.run() to perform the computation.




### **Tensorflow 1.0 Example**

```python
import tensorflow as tf

# Create a graph
g = tf.Graph()

# Define operations (nodes) in the graph
with g.as_default():
    x = tf.placeholder(tf.float32, shape=[None])
    y = tf.constant(5.0)
    z = tf.multiply(x, y)

# Create a session to run the graph
sess = tf.Session(graph=g)

# Feed data to the placeholder and run the computation
feed_dict = {x: [1, 2, 3]}  # Dictionary mapping placeholders to data
output = sess.run(z, feed_dict=feed_dict)
print(output)  # Output: [5. 10. 15.]

# Close the session (recommended)
sess.close()
```

### **Tensorflow 2.0 Example**
* In TensorFlow 2.0, eager execution is enabled by default, meaning operations are executed immediately as they are called in Python, which simplifies model development and debugging.

**Creating and Using Graphs**
* Here's an example of creating and using a graph in TensorFlow 2.0:
* In this example, add is a Python function decorated with tf.function, which makes TensorFlow convert it into a graph. The function can then be called with tensors as arguments, and TensorFlow handles the execution.

```python
import tensorflow as tf

# Define a function and convert it to a graph
@tf.function
def add(a, b):
    return a + b

# Create tensors
a = tf.constant(3)
b = tf.constant(4)

# Call the function
result = add(a, b)
# result: <tf.Tensor: shape=(), dtype=int32, numpy=7>
```

**Using Sessions in TensorFlow 2.0**

* In TensorFlow 2.0, sessions are no longer needed because eager execution is enabled by default. This means you can execute operations immediately, and TensorFlow handles the context and resources internally. However, if you need to run TensorFlow 1.x code, you can still use sessions within a tf.compat.v1 context.
* In this example, eager execution is disabled, and a graph with placeholders is defined. A session is created to run the graph, and the feed_dict parameter is used to pass values to the placeholders.



```python
import tensorflow as tf

# Disable eager execution
tf.compat.v1.disable_eager_execution()

# Define a graph
a = tf.compat.v1.placeholder(tf.int32, shape=())
b = tf.compat.v1.placeholder(tf.int32, shape=())
add_op = a + b

# Create a session and run the graph
with tf.compat.v1.Session() as sess:
    result = sess.run(add_op, feed_dict={a: 3, b: 4})
    # result: 7
```

## **Eager Execution in TensorFlow 2.0**

* Eager execution is an imperative programming environment that evaluates operations immediately, without building graphs. Operations return concrete values instead of constructing a computational graph to be executed later. This makes TensorFlow easier to use and debug, as it behaves more like a standard Python program.

* Eager execution is enabled by default in TensorFlow 2.0, which provides a more interactive experience for developers, simplifies model building, and accelerates debugging.

* There is no need to create sessions or define placeholders. You can use TensorFlow functions directly, and the results are available immediately.



### **Examples**

```python
import tensorflow as tf

# Eager execution is enabled by default in TensorFlow 2.0
# No need to enable it manually

# Define some constants
a = tf.constant(2.0)
b = tf.constant(3.0)

# Perform operations
c = a + b
d = a * b

# The results are available immediately
# c: <tf.Tensor: shape=(), dtype=float32, numpy=5.0>
# d: <tf.Tensor: shape=(), dtype=float32, numpy=6.0>

# Function to perform a simple operation
def simple_function(x, y):
    return tf.add(x, y)

# Call the function
result = simple_function(a, b)
# result: <tf.Tensor: shape=(), dtype=float32, numpy=5.0>
```

### **Advantages of Eager Execution:**

* Interactive Development: Eager execution allows for intuitive, interactive debugging and development.
* Simpler Code: No need to manage sessions or graphs explicitly.
* Immediate Execution: Operations are executed immediately, and results are readily available.
* Integration with Python: Works seamlessly with Python control flow, making it easier to develop and debug models.

## **Broadcasting**

* Broadcasting in TensorFlow allows for efficient element-wise operations on tensors of different shapes. By implicitly expanding the smaller tensor to match the shape of the larger tensor, TensorFlow simplifies the process of performing mathematical operations on tensors. This technique is particularly useful in scientific computing and machine learning, where operations on tensors of varying shapes are common.


## **How to Implement**
* Broadcasting is handled automatically by TensorFlow when performing operations on tensors of different shapes. You do not need to manually broadcast the tensors; TensorFlow takes care of this behind the scenes. You can simply perform operations as you would with tensors of the same shape.

**Example**

```python
import tensorflow as tf

# Define tensors of different shapes
a = tf.constant([1, 2, 3])         # Shape: (3,)
b = tf.constant([[4], [5], [6]])   # Shape: (3, 1)

# Perform element-wise addition with broadcasting
c = a + b

print("\n c : \n ",c)
# The shapes of `a` and `b` are compatible for broadcasting
# `a` is treated as [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
# `b` is treated as [[4, 4, 4], [5, 5, 5], [6, 6, 6]]
# The result `c` is [[5, 6, 7], [6, 7, 8], [7, 8, 9]]
# c: <tf.Tensor: shape=(3, 3), dtype=int32, numpy=
# array([[5, 6, 7],
#        [6, 7, 8],
#        [7, 8, 9]], dtype=int32)>

# Perform element-wise multiplication with broadcasting
d = a * b
print("\n d : \n ",d)

# The shapes of `a` and `b` are compatible for broadcasting
# `a` is treated as [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
# `b` is treated as [[4, 4, 4], [5, 5, 5], [6, 6, 6]]
# The result `d` is [[4, 8, 12], [5, 10, 15], [6, 12, 18]]
# d: <tf.Tensor: shape=(3, 3), dtype=int32, numpy=
# array([[ 4,  8, 12],
#        [ 5, 10, 15],
#        [ 6, 12, 18]], dtype=int32)>
```

**Explanation:**

**Defining Tensors:**

* We define two tensors a and b with shapes (3,) and (3, 1) respectively.

**Element-wise Addition:**

* When performing the addition `a + b`, TensorFlow automatically broadcasts the shapes:
  * `a` is broadcasted to shape (3, 3) by repeating its values across a new axis.
  * `b` is broadcasted to shape (3, 3) by repeating its values along the existing axis.
The result `c` is a tensor with shape (3, 3) where each element is the sum of the corresponding elements in the broadcasted tensors.

**Element-wise Multiplication:**

* Similarly, for the multiplication `a * b`, TensorFlow broadcasts the tensors:
`a` and `b` are broadcasted to shape (3, 3).
* The result `d` is a tensor with shape (3, 3) where each element is the product of the corresponding elements in the broadcasted tensors.


## **Activation Functions**

* Activation functions are mathematical equations that determine the output of a neural network model. **How to Implement**
* In TensorFlow 2.0, you can implement activation functions using built-in functions from the tf.keras.activations module. These functions can be used directly in your model layers or called independently.

```python
import tensorflow as tf
import numpy as np

# Define input tensor
x = tf.constant([-1.0, 0.0, 1.0, 2.0, 3.0])

# Sigmoid Activation
sigmoid_output = tf.keras.activations.sigmoid(x)
# sigmoid_output: <tf.Tensor: shape=(5,), dtype=float32, numpy=array([0.2689414, 0.5, 0.7310586, 0.880797, 0.95257413], dtype=float32)>

# Tanh Activation
tanh_output = tf.keras.activations.tanh(x)
# tanh_output: <tf.Tensor: shape=(5,), dtype=float32, numpy=array([-0.7615942, 0.0, 0.7615942, 0.9640276, 0.9950547], dtype=float32)>

# ReLU Activation
relu_output = tf.keras.activations.relu(x)
# relu_output: <tf.Tensor: shape=(5,), dtype=float32, numpy=array([0., 0., 1., 2., 3.], dtype=float32)>

# Leaky ReLU Activation
leaky_relu_output = tf.keras.layers.LeakyReLU(negative_slope=0.01)(x)
# leaky_relu_output: <tf.Tensor: shape=(5,), dtype=float32, numpy=array([-0.01, 0., 1., 2., 3.], dtype=float32)>

# ELU Activation
elu_output = tf.keras.activations.elu(x)
# elu_output: <tf.Tensor: shape=(5,), dtype=float32, numpy=array([-0.63212055, 0., 1., 2., 3.], dtype=float32)>

# Softmax Activation
x_softmax = tf.constant([1.0, 2.0, 3.0])
softmax_output = tf.keras.activations.softmax(x_softmax)
# softmax_output: <tf.Tensor: shape=(3,), dtype=float32, numpy=array([0.09003057, 0.24472848, 0.66524094], dtype=float32)>
```

**Explanation:**
* **Sigmoid Activation:**

    * Applies the sigmoid function to the input tensor x.

* **Tanh Activation:**

    * Applies the tanh function to the input tensor x.

* **ReLU Activation:**

    * Applies the ReLU function to the input tensor x.

* **Leaky ReLU Activation:**

  * Applies the Leaky ReLU function to the input tensor x with the negative_slope argument set to 0.01, which replaces the deprecated alpha argument.

* **ELU Activation:**

   * Applies the ELU function to the input tensor x.

* **Softmax Activation:**

  * Applies the softmax function to the input tensor x_softmax, converting the values to a probability distribution.
